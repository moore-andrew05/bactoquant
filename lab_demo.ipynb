```python
import numpy as np
from matplotlib import pyplot as plt,cm
```

```python
img = np.random.gamma(1.0, 2.0, (20,100))
img *= 4095 / np.max(img)
img = img.astype("int") 
plt.imshow(img, cmap="gray_r", vmin = 0, vmax = 4095);
```

```python
img = img.max(axis=0)
```

```python
plt.imshow(np.vstack([img for i in range(30)]), cmap="gray_r", vmin = 0, vmax = 4095);
```

```python
img[img<1000] = -1
```

```python
cmap = cm.get_cmap("gray_r")
cmap.set_under("red")
```

```python
plt.imshow(np.vstack([img for i in range(30)]), cmap=cmap, vmin = 0, vmax = 4095);
```

```python
img[img==-1] = 0
img = img/4095
```

```python
def _interp1d(array, new_len):
        la = len(array)
        return np.interp(np.linspace(0, la - 1, num=new_len), np.arange(la), array)
```

```python
img = _interp1d(img, 150)
```

```python
plt.imshow(np.vstack([img for i in range(30)]), cmap=cmap, vmin = 0, vmax = 1.0);
```
